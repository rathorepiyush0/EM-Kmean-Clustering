```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

from sklearn.cluster import KMeans 

#importsome data

from sklearn import datasets

iris = datasets.load_iris()

iris = datasets.load_iris()
x = iris.data
y = iris.target


from pandas.core.arrays.sparse import SparseArray as _SparseArray


x = pd.DataFrame(iris.data)

x.columns = ['sepal_length','sepal_width','petal_length','petal_width']

from pandas.core.arrays.sparse import SparseArray as _SparseArray

y = pd.DataFrame(iris.target)

y.columns = ['Targets']

# build the k means model

model = KMeans(n_clusters=3)

model.fit(x)

## visual the clustering

plt.figure(figsize=(14,14))

colormap = np.array(['red','lime','black'])


#plot the original classification

plt.subplot(2,2,1)

plt.scatter(x.petal_length,x.petal_width,c=colormap[y.Targets],s=40)

plt.title('Real Clusters')

plt.xlabel('petal length')

plt.ylabel('petal width')

# plot the model classification

plt.subplot(2,2,1)

plt.scatter(x.petal_length,x.petal_width,c=colormap[model.labels_],s=40)

plt.title('KMeans  clustering')

plt.xlabel('petal length')

plt.ylabel('petal width')

#General EM for GM


import sklearn.preprocessing as preprocessing
import sklearn.ensemble as ensemble
import sklearn.model_selection as ms

from sklearn import preprocessing

#transform your data

scaler = preprocessing.StandardScaler()

scaler.fit(x)

xsa = scaler.transform(x)

xs = pd.DataFrame(xsa,columns = x.columns)

from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=3)

gm.fit(xs)

gm_y = gm.predict(xs)

plt.subplot(2,2,3)

plt.scatter(x.petal_length,x.petal_width,c=colormap[gm_y],s=40)

plt.title('GM Clustering')

plt.xlabel('petal Length')

plt.ylabel('Petal Width')

print('observation: The GM using EM algorithm based clustering matched the true label more closely than kmeans. ')



```
